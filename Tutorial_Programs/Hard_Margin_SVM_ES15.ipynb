$\text{The process to find a is simple, we apply the ocnstraints and add the regulatization term}$
$\text{The values of Œª range form $10^{-5}$ to $10^5$ (100 values have been taken)}$


---


$\text{Constraints  and Objective:-}$

\begin{align}
      min_{a} \  ||a||_{2} + Œª ||a||_1
      \\
      ‚àÄ x \in X \ \  \ \vec{a} ¬∑ \vec{x}  + b \ge 1
      \\
      ‚àÄ y \in Y \ \  \ \vec{a} ¬∑ \vec{y} + b \le -1
      \\
      T_{max} = max_{Œª} \ \frac{2}{||a||_{2}}
      \\
      C_{min} = min_{Œª} \ (\text{Number of Positions where $a_i \ge 10^{-4}$)}
\end{align}





```python
for lam in lambda_space:
    a = cp.Variable((n,1));
    b = cp.Variable();
    constraints = [a.T@ X[:,0:N] - b >= 1]
    constraints +=[a.T@Y[:,0:M] - b <= -1]
    objective = cp.Minimize(cp.norm(a,2) + lam*cp.norm(a,1))
    prob = cp.Problem(objective,constraints);
    result = prob.solve();
    norm_values.append(2/np.linalg.norm(a.value,2))
    cardinality_values.append(np.sum(np.abs(a.value) > 1e-4))
    if(2/np.linalg.norm(a.value,2) > max_thickness):
        max_thickness = 2/np.linalg.norm(a.value,2)
        max_thickness_cardinality = np.sum(np.abs(a.value) > 1e-4)
        lambda_max = lam
    if((np.sum(np.abs(a.value) > 1e-4)) == 10 and count ==0):
        earliest_10_cardinal  =  a.value;
        earliest_10_features  = np.nonzero(np.abs(a.value) > 1e-4)[0]
        earliest_10_thickness = 2/np.linalg.norm(a.value)
        count = count + 1


```
